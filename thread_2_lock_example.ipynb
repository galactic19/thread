### thread_2_lock 파일의 예제를 기반으로 새로운 코드를 직접 작성 해보기.


```python
import threading
import time

lock = threading.Lock()

global global_list
global_list = []


def add_odd(*args, **kwargs):
    for _ in args[0]:
        if _ // 2:
            with lock:
                global_list.remove(_)
                
def add_even(*args, **kwargs):
    for _ in args[0]:
        if _ // 3:
            with lock:
                global_list.remove(_)
                
def list_manager(*args, **kwargs):
    for _ in range(args[0]):
        with lock:
            global_list.append(_)
                

thread3 = threading.Thread(target=list_manager, args=(10000,))
thread1 = threading.Thread(target=add_odd, args=(global_list,))
thread2 = threading.Thread(target=add_even, args=(global_list,))


thread3.start()
thread1.start()
thread2.start()

thread3.join()
thread1.join()
thread2.join()

print(global_list)
print(len(global_list))
```
